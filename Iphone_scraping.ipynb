```python
import requests
from bs4 import BeautifulSoup

# Specify the URL of the website to scrape
url = "https://www.flipkart.com/search?q=iphone"

# Send a GET request to the specified URL and get the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(html_content, "html.parser")

# Find all the products in the search results
products = soup.find_all("div", {"class": "_2kHMtA"})

# Loop through each product and extract the relevant information
for product in products:
    # Get the name of the product
    name = product.find("div", {"class": "_4rR01T"}).text.strip()
    # Get the price of the product
    price = product.find("div", {"class": "_30jeq3 _1_WHN1"}).text.strip()
    # Get the rating of the product
    rating = product.find("div", {"class": "_3LWZlK"}).text.strip()
    
    # Print the extracted information
    print("Name: ", name)
    print("Price: ", price)
    print("Rating: ", rating)
    print("\n")
```
